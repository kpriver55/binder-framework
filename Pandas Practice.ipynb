```python
energy['domain'] = energy['company email'].str.split('@')
```

```python
energy['domain'].head()
```

```python
energy_domain = energy.dropna(subset = ['domain'], axis = 0)
```

```python
energy_domain.head()
```

```python
domain_names = []
domains = energy_domain['domain'].values.tolist()
print(domains[0])
```

```python
print(domains[72])
```

```python
domains[72] = ['N/A', 'billieannplastics.com']
```

```python
for i in range(len(domains)):
    print(len(domains[i]))
```

It seems that some of our email addresses did not include both an address and a domain name, we'll need to search for these entries and either correct them or drop them. The evidence for this is that not all members of domains list have length 2. Those with length one are missing one of the two necessary components.

```python
for i in range(len(domains)):
    if len(domains[i]) == 1:
        print(domains[i])
    
```

Observing the divergent examples, we find that each of them are missing the address component instead of the domain name. We can fix this by adding an arbitrary first list element to each as we only care about the domain name.

```python
for i in range(len(domains)):
    if len(domains[i]) == 1:
        domains[i].insert(0, 'N/A')
        
```

Now, we can extract the domain names from the list of lists without incident

```python
for item in domains:
    domain_names.append(item[1])
    
```

```python
from collections import Counter
```

```python
unique_domain_names_with_count = Counter(domain_names)
```

The above returns a counter type object which is similar to a dictionary giving the unique domain names as keys and the number of times they appear as values. The number of keys in the number of unique domain names

```python
number_of_domain_names = len(unique_domain_names_with_count.keys())
print(number_of_domain_names)
```

The number of unique domain names given in the email address section is 361

## Question 4

```python
energy.dropna(axis = 0, subset = ['NTA'], inplace = True)
```

```python
energy_NTA = energy.groupby(['NTA'])
```

```python
energy_NTA = energy_NTA.filter(lambda x : len(x) >= 5 )
```

The above function filters out NTAs which have less than five instances in the data set as demonstrated by the value counts table below

```python
energy_NTA['NTA'].value_counts()
```

```python
energy_NTA.head()
```

We have a problem with null values in the job created column. If no value is reported, it is reasonable to expect that that value will be zero, so we will fill them as such

```python
values = {'Job created' : 0}
energy_NTA.fillna(value = values, inplace = True)
```

```python
mean_savings_jobs_by_NTA = energy_NTA.groupby(['NTA'])[['Total Savings', 'Job created']].mean()
mean_savings_jobs_by_NTA['Job created'] = energy_NTA.groupby(['NTA'])['Job created'].sum()
mapper = {'Total Savings' : 'Average Savings'}

mean_savings_jobs_by_NTA = mean_savings_jobs_by_NTA.rename(columns = mapper)
mean_savings_jobs_by_NTA.head(20)
```

The above displays the average total savings and total jobs created for companies within each NTA. It is worth noting that the amount of missing data in the jobs created column does make the results for this parameter unreliable

```python
mean_savings_jobs_by_NTA.to_csv('jobs_and_savings.csv')
```

For the last part, we'll need to get the original data set back, as we had dropped some entries on the basis that the NTA column had no value. We now want to have those data points back, so we'll re-read the original csv into a new dataframe.

```python
energy_2 = pd.read_csv('Energy_Cost_Savings.csv')
energy_2['Effective Date'].isna().sum()
```

```python
import datetime
months = []
for entry in energy_2['Effective Date']:
    datee = datetime.datetime.strptime(entry, "%m/%d/%Y")
    month = datee.month
    months.append(month)
    
print(len(months))


    
```

```python
energy_2['month'] = pd.Series(months)
```

```python
total_jobs_by_month = energy_2.groupby(['month'])['Job created'].sum()
```

```python
total_jobs_by_month.head(12)
```

```python
plt.plot(total_jobs_by_month.keys(), total_jobs_by_month)
plt.xlabel('Month (by number)')
plt.ylabel('Total Jobs Created')
plt.title('Jobs Created By Month')
plt.show()
```
